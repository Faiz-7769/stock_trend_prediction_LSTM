# **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

# **Data Exploration**

```python
data = yf.Ticker("MSFT")
```

```python
msft = data.history(period="max")
msft
```

```python
msft.index
```

```python
msft.plot.line(y="Close", use_index = True)
```

**Target Creation**

```python
msft["Tomorrow"] = msft["Close"].shift(-1)
msft
```

# **Scaling And Splitting**

```python
total_length = len(msft)
train_dev_split_index = int(total_length * 0.8)
dev_test_split_index = int(total_length * 0.9)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
msft_scaled = scaler.fit_transform(msft)
```

```python
# Convert the scaled NumPy array back to a pandas DataFrame
msft_scaled_df = pd.DataFrame(msft_scaled, columns=msft.columns, index=msft.index)

train_df = msft_scaled_df.iloc[:train_dev_split_index]
dev_df = msft_scaled_df.iloc[train_dev_split_index:dev_test_split_index]
test_df = msft_scaled_df.iloc[dev_test_split_index:]
plt.plot(train_df["Close"], label="train")
plt.plot(dev_df["Close"], label="dev")
plt.plot(test_df["Close"], label="test")
plt.legend(['Train',"Validation","Test"])
```

**Feature Seelction**

```python
features = ['Open', 'High', 'Low', 'Close', 'Volume']
target = 'Tomorrow'
X = msft[features]
Y = msft[target]

scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_x.fit_transform(X)
Y_scaled = scaler_y.fit_transform(Y.values.reshape(-1,1))


x_df = pd.DataFrame(X_scaled, columns=features,index = X.index)
y_df = pd.DataFrame(Y_scaled, columns=[target],index = Y.index)

train_x_df = x_df.iloc[:train_dev_split_index]
dev_x_df = x_df.iloc[train_dev_split_index:dev_test_split_index]
test_x_df = x_df.iloc[dev_test_split_index:]


X_train = train_x_df[features]
y_train = y_df.iloc[:train_dev_split_index]

X_dev = dev_x_df[features]
y_dev = y_df.iloc[train_dev_split_index:dev_test_split_index]

X_test = test_x_df[features]
y_test = y_df[dev_test_split_index:]
```

# **Model Using LSTM**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping


model = Sequential([layers.Input((5,X_train.shape[1])),
                    layers.LSTM(128,return_sequences=True),
                    layers.LSTM(64),
                    layers.Dropout(0.2),
                    layers.Dense(32, activation = 'relu'),
                    layers.Dropout(0.2),
                    layers.Dense(32,activation = 'relu'),
                    layers.Dense(1)])

early_stop = EarlyStopping(
    monitor='val_loss', patience=10, restore_best_weights=True)

model.compile(loss="mse", optimizer=Adam(learning_rate=0.001),metrics = ['mean_absolute_error'])

# Reshape input data for LSTM [samples, timesteps, features]
def create_sequences(X, y, window_size=5):
    X_seq = []
    y_seq = []
    for i in range(window_size, len(X)):
        X_seq.append(X.iloc[i-window_size:i].values)  # (window_size, features)
        y_seq.append(y.iloc[i])                       # Predict the next value
    return np.array(X_seq), np.array(y_seq)

X_train_reshaped, y_train_seq = create_sequences(X_train, y_train, window_size=5)
X_dev_reshaped, y_dev_seq     = create_sequences(X_dev, y_dev, window_size=5)
X_test_reshaped, y_test_seq   = create_sequences(X_test, y_test, window_size=5)

model.fit(X_train_reshaped, y_train_seq, validation_data=(X_dev_reshaped, y_dev_seq), epochs=50,callbacks=[early_stop],batch_size= 32,verbose = 1)
```

# **Prediction**

```python
train_pred = model.predict(X_train_reshaped).flatten()
dev_pred = model.predict(X_dev_reshaped).flatten()
test_pred = model.predict(X_test_reshaped).flatten()
```

# **Observed VS Predicted**

```python
plt.plot(train_df.index[5:], train_pred, label='Training Predictions')
plt.plot(train_df.index[5:], y_train_seq, label='Training Observations')
plt.legend()
plt.show()
```

```python
plt.plot(dev_df.index[5:], dev_pred, label='Training Predictions')
plt.plot(dev_df.index[5:], y_dev_seq, label='Training Observations')
plt.legend()
plt.show()
```

```python
plt.plot(test_df.index[5:], test_pred, label='Training Predictions')
plt.plot(test_df.index[5:], y_test_seq, label='Training Observations')
plt.legend()
plt.show()
```

```python
# Inverse transform predictions
train_pred = scaler_y.inverse_transform(train_pred.reshape(-1,1)).flatten()
dev_pred   = scaler_y.inverse_transform(dev_pred.reshape(-1, 1)).flatten()
test_pred  = scaler_y.inverse_transform(test_pred.reshape(-1, 1)).flatten()

# Inverse transform actual values
y_train_seq = scaler_y.inverse_transform(y_train_seq.reshape(-1, 1)).flatten()
y_dev_seq   = scaler_y.inverse_transform(y_dev_seq.reshape(-1, 1)).flatten()
y_test_seq  = scaler_y.inverse_transform(y_test_seq.reshape(-1, 1)).flatten()
```

```python
window_size = 5  # same as used in your create_sequences()

plt.figure(figsize=(14, 6))

# Plot Training Data
plt.plot(train_df.index[window_size:], train_pred, label='Training Predictions', color='blue')
plt.plot(train_df.index[window_size:], y_train_seq, label='Training Observations', color='cyan')

# Plot Dev Data
plt.plot(dev_df.index[window_size:], dev_pred, label='Dev Predictions', color='green')
plt.plot(dev_df.index[window_size:], y_dev_seq, label='Dev Observations', color='lightgreen')

# Plot Test Data
plt.plot(test_df.index[window_size:], test_pred, label='Test Predictions', color='red')
plt.plot(test_df.index[window_size:], y_test_seq, label='Test Observations', color='salmon')

# Final touches
plt.xlabel("Time")
plt.ylabel("Target Value")
plt.title("Predictions vs. Observations")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
